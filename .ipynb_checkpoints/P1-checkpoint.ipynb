Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math
# Needed for average_lines function
from scipy.cluster.hierarchy import fcluster, linkage

def read_image(file):
    return mpimg.imread(file)

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_noise(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def define_region_of_interest(img):
    """
    Determines the vertices of a triangle.
    
    """
    imshape = img.shape
    # place base of triangle at bottom 1/6 way in to the left and  1/8 right
    # add vertex at 1/2 across and 1/2 down
    return np.array([[(round(imshape[1]/6),imshape[0]), (imshape[1]/2, imshape[0]/2), (round(7*imshape[1]/8),imshape[0])]], dtype=np.int32)


def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def find_line(x1, y1, x2, y2):
    # check for vertical slope
    if (x2-x1) == 0:
        # make slope some non-infinite but high number
        m = 100
    else :
        m = (y2 - y1)/(x2 - x1)
    b = y1 - m*x1
    return m, b

def get_lane_pts(imshape, m, b):
    
    # find points of line that intersect bottom of image
    # y = height = mx + b => x = (height - b)/m
    x_bottom = round((imshape[0] - b)/m)
        
    # find upper point where y = 3/5 image height
    # y = 3*height/5 = mx + b => x = (3*height/5 - b)/m
    x_top = round((3*imshape[0]/5 - b)/m)
    return (int(x_bottom), imshape[0]), (int(x_top), int(round(3*imshape[0]/5)))
# We expect lanes to not be horizontal!!
# Therefore we can throw out any lines that are close to horizontal
def slope_check(m):
    if m == 0.:
        return False
    else:
        return True
# Use hierarchial clustering to find groups of similar slope, then find their mean
# We expect one lane to have positive slope and the other negative slope
def average_lines(slopes, intercepts):
    positive_m = []
    positive_b = []
    negative_m = []
    negative_b = []
    for index in range(len(slopes)):
        if slopes[index] > 0:
            positive_m.append(slopes[index])
            positive_b.append(intercepts[index])
        else:
            negative_m.append(slopes[index])
            negative_b.append(intercepts[index])
    
    return np.mean(positive_m), np.mean(positive_b), np.mean(negative_m), np.mean(negative_b)

def draw_lines(img, lines, color=[255, 0, 0], thickness=8):
    """
    Draws `lines` with `color` and `thickness`.
    
    Lines are drawn on the image inplace (mutates the image).
    """
    slopes = []
    intercepts = []
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                m, b = find_line(x1, y1, x2, y2)
                if slope_check(m):
                    slopes.append(m)
                    intercepts.append(b)
    m1, b1, m2, b2 = average_lines(slopes, intercepts)
    slope1 = np.mean(m1)
    intercept1 = np.mean(b1)
    slope2 = np.mean(m2)
    intercept2 = np.mean(b2)
    if not math.isnan(slope1):
        pt1, pt2 = get_lane_pts(img.shape, slope1, intercept1)
        cv2.line(img, pt1, pt2, color, thickness)
    if not math.isnan(slope2):
        pt1, pt2 = get_lane_pts(img.shape, slope2, intercept2)
        cv2.line(img, pt1, pt2, color, thickness)
                
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros(img.shape, dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def stack_images(img):
    return np.dstack((img, np.zeros(img.shape, dtype=np.uint8), np.zeros(img.shape, dtype=np.uint8))) 

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
```

```python
# process_image function
# input is color image
def process_image(image):
    # convert to grayscale
    gray_img = grayscale(image)
    # apply gaussian noise
    smooth = gaussian_noise(gray_img, 7)
    # apply Canny edge detection
    edges = canny(smooth, 50, 130)
    # define area of interest
    # NOTE: I added this helper function, see above
    vertices = define_region_of_interest(edges)
    # get masked img
    mask_img = region_of_interest(edges, vertices)
    # apply Hough transform
    rho = 1
    theta = np.pi/180
    lane_img = hough_lines(mask_img, rho, theta, 1, 20, 1)
    # combine drawn lanes with original image
    result = weighted_img(stack_images(lane_img), image)
    return result
```

My Reflections
-----------
Static algorithms such as the pipeline I created above are highly susceptible to environmental conditions. For example, the lighting has a huge impact on this pipeline as the thresholds for the Canny edge detection algorithm would vary greatly between a dark image and a bright image. Without some way to normalize the images to account for differences in illumination, my pipeline will perform well for some images and poorly for others. A more robust algorithm might preprocess all the images to attempt to calibrate either the RGB or grayscale image. Then, parameters such as those used for Canny edge detecting, can be more universally applicable.

Another shortcoming of this pipeline is the core assumption that lanes will be straight and that each lane segment is of a certain length. Given a road with curvy lanes or lanes that are not visible because of some obstruction, the algorithm will either fail to find the lanes or will project lanes in a direction that is not accurate. A method to improve the algorithm would be to either identify lanes and keep the lane projections close to the car to avoid missing curves, or the lane finding algorithm could search for polynomial fits to the edges instead of only fitting straight lines. Another potential fix is a recursive algorithm that takes the last known lane positions and uses those as the starting point for the next image in the sequence.

Overall, I advise that for any algorithm that seeks to sense the real world, a combination of several types of sensors: optical, audio, etc., should be used to remedy the shortcomings of each individual sensor.

```python
# perform adaptive histogram equalization to image
# input img is full color RGB image
def hist_eq(img):
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cll_red = clahe.apply(img[:,:,0])
    cll_blue = clahe.apply(img[:,:,1])
    cll_green = clahe.apply(img[:,:,2])
    return np.dstack((cll_red, cll_blue, cll_green))
# define color boundaries for yellow and white
def color_boundaries():
    # NOTE: colors are in RGB format
    # lower and upper RGB limits for yellow
    yellow = ([200,200,0], [255,255,200])
    # lower and upper RGB limits for white
    white = ([215,215,215], [255,255,255])
    # combine
    return [yellow, white]
# find colors in HSV image
def color_mask(img):
    # create empty copy of img
    masks = np.copy(img[:,:,0])*0
    # get hsv color bounds
    boundaries = color_boundaries()
    for color in boundaries:
        # Threshold the image
        lower = color[0]
        upper = color[1]
        mask = cv2.inRange(img, np.array(lower, dtype="uint8"), np.array(upper, dtype="uint8"))
        masks = weighted_img(mask, masks, =1., =1., =0.)
    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(img, img, mask= masks)
    return res
# convert to binary image
def to_binary(img, lower_thresh, upper_thres):
    ret,thresh = cv2.threshold(img, lower_thresh, upper_thres, 0)
    return thresh
# find contours
# img is a binary image
def find_contours(img):
    image, contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    return contours
# img is full color image
def draw_contours(img, contours):
    return cv2.drawContours(img, contours, -1, (0,255,0), 3)
```

```python
# process_image_challenge function
# input is color image
# NOTE: Tried to use regular canny edge detection but was not satisfied
# Switched to finding image contours and got slightly better results
def process_image_challenge(image):
    # equalize image
    eq_image = hist_eq(image)
    # apply gaussian noise
    smooth = gaussian_noise(eq_image, 7)
    # get region of interest
    vertices = define_region_of_interest(smooth)
    roi = region_of_interest(smooth, vertices)
    # apply color thresholding 
    color_thresh = color_mask(roi)
    # convert to grayscale
    gray_img = grayscale(color_thresh)
    # convert to binary
    binary_img = to_binary(gray_img, 50, 255)
#     SECTION OF CODE FOR ONLY CANNY
#     # apply Canny edge detection
#     edges = canny(gray_img, 50, 220)
#     # define area of interest
#     # NOTE: I added this helper function, see above
#     vertices = define_region_of_interest(edges)
#     # get masked img
#     mask_img = region_of_interest(edges, vertices)
    # find contours and apply canny to contours
    contours = find_contours(binary_img)
    # create empty image
    img_copy = np.copy(image)*0
    # draw contours
    drawn = draw_contours(img_copy, contours)
    # apply edge detection
    edges = canny(grayscale(drawn), 50, 220)
    # apply Hough transform
    rho = 1
    theta = np.pi/180
    lane_img = hough_lines(edges, rho, theta, 1, 5, 1)
    # combine drawn lanes with original image
    result = weighted_img(stack_images(lane_img), image)
    return result
```
